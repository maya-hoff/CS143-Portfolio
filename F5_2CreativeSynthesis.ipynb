```python
import nltk
from nltk.corpus import wordnet as wn
import random
nltk.download('wordnet')
```

```python
from google.colab import files

uploaded = files.upload()
filename = next(iter(uploaded))

with open(filename, "r") as f:
    words_list = f.read().splitlines()
```

```python
def get_similarity(word1, word2):
  synsets1 = wn.synsets(word1)
  synsets2 = wn.synsets(word2)
  if not synsets1 or not synsets2:
    return 0
  similarity = synsets1[0].wup_similarity(synsets2[0])
  if similarity is None:
    return 0

  return round(similarity * 100, 2)

def generate_words():
  return random.sample(words_list, 2)

def game_loop():
  word1, word2 = generate_words()
  while True:
    print(f"\nHow similar are these words?")
    print(f"{word1} --- {word2}")

    user_input = input("Enter similarity score (0-100): ")

    if user_input.strip() == "":
      print("Please enter a number.")
      continue

    try:
      guess = float(user_input)
      if guess < 0 or guess > 100:
        print("Please enter a number between 1 and 100.")
        continue

      break

    except ValueError:
      print("Invalid input.")
      continue

  actual = get_similarity(word1, word2)
  difference = abs(actual - guess)
  print(f"\nActual similarity: {actual}")

  if difference <= 10:
    print("Excellent guess!")
    return 1
  elif difference <= 25:
    print("Pretty close!")
    return 0.5
  else:
    print("Not very close.")
    return 0
```

```python
score = 0

for i in range(5):
    result = game_loop()
    score += result
    print("Current score:", score)

print("\nFinal score:", score)
```
