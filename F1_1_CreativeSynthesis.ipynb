```python
from google.colab import files

uploaded = files.upload()
filename = next(iter(uploaded))

with open(filename, "r") as f:
    sentence_list = f.read().splitlines()
```

```python
from transformers import pipeline
from accelerate import Accelerator

device = Accelerator().device

classifier = pipeline("sentiment-analysis", model="SamLowe/roberta-base-go_emotions", device=device)
```

```python
import random

def classify(sentence):
  result = classifier(sentence)
  result_cleaned = result[0]["label"]
  return result_cleaned

def get_sentence():
  return random.sample(sentence_list, 1)
```

```python
possible_emotions = [
    "admiration",
    "amusement",
    "anger",
    "annoyance",
    "approval",
    "caring",
    "confusion",
    "curiosity",
    "desire",
    "disappointment",
    "disapproval",
    "disgust",
    "embarrassment",
    "excitement",
    "fear",
    "gratitude",
    "grief",
    "joy",
    "love",
    "nervousness",
    "neutral",
    "optimism",
    "pride",
    "realization",
    "relief",
    "remorse",
    "sadness",
    "surprise"
]
```

```python
def game_loop():
  curr_sentence = get_sentence()
  while True:
    print(curr_sentence)
    user_input = input("Guess the sentence emotion: ").lower().strip()

    if user_input == "":
      print("Please enter a valid emotion.")
      continue

    if user_input not in possible_emotions:
      print("Please enter a valid emotion.")
      continue

    guess = user_input
    break

  actual = classify(curr_sentence).lower()
  print(f"\nActual emotion: {actual}")
  if actual == guess:
    print("You're right!")
    return 1

  else:
    print("Incorrect.")
    return 0
```

```python
score = 0

for i in range(5):
    result = game_loop()
    score += result
    print("Current score:", score)

print("\nFinal score:", score)
```
